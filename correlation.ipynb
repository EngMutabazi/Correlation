```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_excel('Production.xlsx')
data
```

```python
ax=data.plot(x='Year',y=['Production(T)'],kind='bar',figsize=(18,8), color='green')
width_scale = 0.60
plt.legend(loc='upper left')
plt.xlabel('Year', fontsize = 15, weight = 'bold')
plt.ylabel('Green Leaf Tea Production(T)', fontsize = 14, weight = 'bold')
ax2 = ax.twinx()
data.plot(y='Fertilizer(T)',kind='line',color= '0',ax=ax2, marker="*")
# ax2.set_ylim(15,19)
plt.title('prod&fertilizer', fontsize = 14, weight = 'bold')
plt.ylabel('Fertilizer(T)', fontsize = 14, weight = 'bold')
plt.legend(loc='upper right')
plt.gcf().axes[0].yaxis.get_major_formatter().set_scientific(False)
plt.ion()
plt.grid()
# plt.savefig("green1.png")
plt.show()
```

```python
ax=data.plot(x='Year',y=['Production(T)', 'Fertilizer(T)'],kind='bar',figsize=(18,8))
width_scale = 0.60
plt.legend(loc='upper left')
plt.xlabel('Year', fontsize = 15, weight = 'bold')
plt.gcf().axes[0].yaxis.get_major_formatter().set_scientific(False)
plt.title('Variation of Coffee production & Fertilizer 2013-2021', fontsize = 14, weight = 'bold')
plt.ylabel('Green Leaf Tea Production(T) & Frtilizer(T)', fontsize = 14, weight = 'bold')
plt.legend(loc='upper right')
plt.ion()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(12,4))
sns.regplot(data["Fertilizer(T)"], data["Production(T)"],ci=None)
plt.gcf().axes[0].yaxis.get_major_formatter().set_scientific(False)
plt.gcf().axes[0].xaxis.get_major_formatter().set_scientific(False)
plt.title("Coffee production vs fertilizer 2013-2021")
print("r=", data['Production(T)'].corr(data['Fertilizer(T)']))
```
